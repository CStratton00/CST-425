## Demonstrate Linear Modeling and finding Maximum Likelihood
Linear models are used to predict the value of a variable based on the values of other variables. One such method is the Maximum Likelihood Estimator. The Maximum Likelihood Estimator is a method for estimating the parameters of a linear model. It is used to find the best fit line for a set of data.

The below code uses a wine dataset to demonstrate Linear Modeling and Maximum Likelihood Estimation.

```python
# import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read in the wine dataset
df = pd.read_csv('winequality-white.csv', sep=';')
df.head()
```

```python
# create a heat map of the correlation matrix
sns.heatmap(df.corr())
plt.show()
```

```python
# based on the heatmap, we can see that residual sugar and density are highly correlated.
# we can create a linear model using the residual sugar and density as a predictor for quality
sns.regplot(x=df['residual sugar'], y=df['density'], data=df)
plt.show()
```

```python
# furthermore, we can use a histogram and density plot to find the maximum likelihood of the quality
sns.distplot(df['quality'], hist=True, kde=True, color='darkblue', hist_kws={'edgecolor':'black'})
plt.show()
```
